```python
df.drop(0, inplace=True)
```

```python
df.insert(0, 'id', range(1, 3))
df.reset_index(drop=True)
```

```python
df_melt = pd.melt(df, id_vars=['id','gov_initiative'], var_name='annee', value_name='valeur')
df_melt 
```

```python
df_melt ['id'] = range(1, len(df_melt ) + 1)
df_melt
```

```python
df_melt.dtypes
```

```python
from unidecode import unidecode
df_melt['gov_initiative'] = df_melt['gov_initiative'].apply(unidecode)
df_melt
```

```python
df_melt.to_csv('gov_ev_initiatives_budget.csv', index=False)
```

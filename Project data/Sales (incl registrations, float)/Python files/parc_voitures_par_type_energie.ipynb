```python
import pandas as pd
import numpy as np
df = pd.read_csv('parc_voitures_par_type_energie.csv')
df
```

```python
df.dtypes
```

```python
df = df.rename(columns={'Type energie':'type_energie','Type voiture':'type_voiture'})
```

```python
df.shape
```

```python
melted_df = pd.melt(df, id_vars=['id', 'type_energie', 'type_voiture'], 
                       var_name='annee', value_name='valeur')
melted_df
```

```python
melted_df['id'] = range(1, len(melted_df) + 1)
melted_df
```

```python
#for column in df.select_dtypes(include=['float64']).columns:
#df[column] = df[column].fillna(0).astype(int)
```

```python
from unidecode import unidecode
melted_df['type_energie'] = melted_df['type_energie'].apply(unidecode)
melted_df['type_voiture'] = melted_df['type_voiture'].apply(unidecode)
```

```python
melted_df.to_csv('parc_voitures_par_type_energie_ev', index=False)
```

```python
import pandas as pd
from unidecode import unidecode

df = pd.read_csv('immatriculations _voitures_ev_neuves_partype_10ans.csv')
df
```

```python
df = df.rename(columns={"Type d'énergie":'type_energie'})
```

```python
df.dtypes
```

```python
df.insert(0, 'id', range(1, 7))
df.reset_index(drop=True)
```

```python
df_melt = pd.melt(df, id_vars=['id','type_energie'], var_name='annee', value_name='valeur')
df_melt 
```

```python
df_melt ['id'] = range(1, len(df_melt ) + 1)
df_melt
```

```python
df_melt['type_energie'] = df_melt['type_energie'].apply(unidecode)
df_melt
```

```python
df_melt.to_csv('immatriculations _voitures_ev_neuves_partype_10ans.csv', index=False)
```

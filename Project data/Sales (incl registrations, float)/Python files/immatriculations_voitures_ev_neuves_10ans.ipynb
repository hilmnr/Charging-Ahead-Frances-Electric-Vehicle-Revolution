```python
import pandas as pd

df = pd.read_csv('immatriculations_voitures_ev_neuves_10ans.csv')
df
```

```python
for col in df.columns:
    if col != 'mois':  
        if df[col].dtype == 'object' and df[col].str.contains(',').any():
            df[col] = df[col].str.replace(',', '').astype(int)
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.insert(0, 'id', range(1, 14))
```

```python
df = df.drop(index=12)
df
```

```python
df_melted = pd.melt(df, id_vars=['id', 'mois'], var_name='annee', value_name='valeur')
df_melted.reset_index
df_melted
```

```python
df_melted['id'] = range(1, len(df_melted) + 1)
```

```python
df_melted
```

```python
df_melted.to_csv('immatriculations_voitures_ev_neuves10ans.csv', index=False)
```
